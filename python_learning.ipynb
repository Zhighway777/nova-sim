```python
import os

# In a Jupyter notebook __file__ is not defined. Fall back to the current working directory.
if '__file__' in globals():
	file_path = __file__
	base_dir = os.path.dirname(os.path.realpath(file_path))
else:
	# Running interactively (notebook); use current working directory as the "file" location.
	base_dir = os.getcwd()
	file_path = os.path.join(base_dir, '<notebook>')

print(file_path)
print(os.path.join(base_dir, '..'))
print(base_dir)
print(os.path.abspath(base_dir))
```
